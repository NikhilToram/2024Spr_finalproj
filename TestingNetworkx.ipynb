```python
import networkx as nx
import matplotlib.pyplot as plt

# Number of rows and columns in the grid
rows = 2
cols = 3

# Number of outer nodes (points of the star)
n = 5

# Create an empty graph
grid = nx.Graph()

# Create pointed star lattices and add them to the grid
for i in range(rows):
    for j in range(cols):
        # Create a complete graph for the outer nodes
        outer_star = nx.complete_graph(n)

        # Create a star graph for the central hub
        hub = nx.star_graph([0])

        # Combine the hub and the outer nodes
        pointed_star = nx.compose(hub, outer_star)

        # Position the pointed star lattice in the grid
        for node in pointed_star.nodes():
            if node == 0:
                # Manually set position for central hub node
                grid.add_node(node, pos=(i * (n + 1), j * (n + 1)))
            else:
                # Position outer nodes relative to the hub
                grid.add_node(node, pos=(i * (n + 1) + node, j * (n + 1)))

        # Add edges to the grid
        grid.add_edges_from(pointed_star.edges())

# Draw the graph with specified positions for all nodes
pos = nx.get_node_attributes(grid, 'pos')
nx.draw(grid, pos, with_labels=True, node_color='skyblue', node_size=1000, font_size=12)
plt.title("Grid of Pointed Star Lattices")
plt.show()
```
